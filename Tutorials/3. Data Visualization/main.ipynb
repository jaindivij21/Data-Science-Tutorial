```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plot

```python
# changing graph theme
themes = plt.style.available
print(themes)
# use seaborn
plt.style.use("seaborn-poster")
```

```python
# defining the plotting coordinates
x = np.arange(10)
y1 = x ** 2
y2 = 2 * x + 3

# plotting the graph
plt.plot(x, y1, 'ro--', label="Apple")
plt.plot(x, y2, color="blue", label="Kiwi")

# extra information
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of fruit over time")
plt.legend()

# shows the graph object on the output screen
plt.show()

# NOTE:
# 1) After you print the graph using plt.show(), the program assumes you have complteted the use of graph; so any more
#    points that you try to print will be shown on a different graph.
# 2) If you pass only one array to the plot function, then it considers x array values as 0, 1, 2 ... while the passed
#    array values as y axis.
```

```python
# documentation
plt.plot?
```

## Scatter Plots


```python
# adjust the size of the plot
plt.figure(figsize=(10, 5), facecolor="grey")

# plotting the graph
plt.scatter(x, y1, color="red", label="Apple", marker='o')
plt.scatter(x, y2, color="green", label="Kiwi", marker='*')

# extra information
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot: Prices of fruit over time")
plt.legend()

# shows the graph object on the output screen
plt.show()
```

## Bar Graph


```python
# documentation
plt.bar?
```

```python
# change theme
plt.style.use("dark_background")

# Example bar graph to show comparison in prices of metals
x_coordinates = np.array([0, 1, 2])
plt.bar(x_coordinates, [30000, 10000, 2000], width=-0.3, align="edge", label="Previous Year",
        tick_label=['Gold', 'Silver', 'Iron'], color="yellow")  # previous year
plt.bar(x_coordinates, [42000, 16000, 3500], width=0.3, align="edge", label="Current Year",
        tick_label=['Gold', 'Silver', 'Iron'], color="grey")  # current year
plt.legend()
plt.ylim(0, 100000)
plt.title("Prices of metals over 2 years")
plt.xlabel("Metals")
plt.ylabel("Prices")
plt.show()
```

## Pie Charts

```python
# documentation
plt.pie?
```

```python
# example of pie chart to represent country and populations

# information
countries = ["India", "China", "USA", "Russia", "France", "UK"]
population = [1.2, 1.3, 0.6, 0.3, 0.2, 0.08]

# formatting
colours = [None, 'yellow', None, None, None, None]

# plotting
plt.pie(population, labels=countries, explode=[.1, 0, 0, 0, 0, 0], shadow='true', autopct='%1.1f%%',
        textprops={'color': 'grey', 'fontsize': 15})  # explode: pass in the radius of the offset
plt.title("Populations")
plt.show()

```
