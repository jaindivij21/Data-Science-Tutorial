```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import all the 5 images
btleft = cv2.imread('files/bottom_left.jpg')
btright = cv2.imread('files/bottom_right.jpg')
center = cv2.imread('files/center.jpeg')
tpleft = cv2.imread('files/top_left.jpg')
tpright = cv2.imread('files/top_right.jpg')
```

```python
# color them to rgb and resize all of them to 200 px
bottomleft = cv2.cvtColor(btleft, cv2.COLOR_BGR2RGB)
bottomright = cv2.cvtColor(btright, cv2.COLOR_BGR2RGB)
centre = cv2.cvtColor(center, cv2.COLOR_BGR2RGB)
topleft = cv2.cvtColor(tpleft, cv2.COLOR_BGR2RGB)
topright = cv2.cvtColor(tpright, cv2.COLOR_BGR2RGB)
```

```python
bottomleft = cv2.resize(bottomleft, (200, 200))
bottomright= cv2.resize(bottomright, (200, 200))
centre = cv2.resize(centre, (100, 100))
topleft = cv2.resize(topleft, (200, 200))
topright = cv2.resize(topright, (200, 200))
```

```python
# adding borders
bl = cv2.copyMakeBorder(bottomleft,10, 10, 10, 10,cv2.BORDER_CONSTANT)
tl = cv2.copyMakeBorder(topleft, 10, 0, 10, 10, cv2.BORDER_CONSTANT)
c = cv2.copyMakeBorder(centre, 10, 10, 10, 10, cv2.BORDER_CONSTANT)
br = cv2.copyMakeBorder(bottomright, 10, 10, 0, 10, cv2.BORDER_CONSTANT)
tr = cv2.copyMakeBorder(topright, 10, 0, 0, 10, cv2.BORDER_CONSTANT)
```

```python
# test print
plt.imshow(bl)
plt.title(bl.shape)
plt.axis("off")
plt.show()
```

```python
top = np.hstack((tl, tr))
bottom = np.hstack((bl, br))
final = np.vstack((top, bottom))
plt.imshow(final)
plt.title(final.shape)
plt.style.use('seaborn')
plt.show()
```

```python
# insert center image
plt.imshow(c)
plt.title(c.shape)
plt.show()
plt.imshow(final)
plt.show()
```

```python
# slicing
final[155:275,155:275,:] = c
plt.imshow(final)
plt.show()
```

```python
final = final.reshape(-1,3)
df = pd.DataFrame(final, columns = ['r', 'g', 'b'])
df.head()
```
