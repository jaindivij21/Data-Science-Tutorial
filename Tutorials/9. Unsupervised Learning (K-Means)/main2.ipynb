### K-Means++

- K-Means is quite sensitive to initalization, if the init is not good, our algorithm is not able to make desired number
of clusters.
- To overcome this problem, we use technique called K-Means++ (described in paper [Robust Seed Selection for K-Means type of Algorithms](https://arxiv.org/abs/1202.1585))
    which chooses initial centers so that they are statiscallly close to final ones.

- The method is default choice for sklearn, so we will use it directly without going into much maths!


Two different types of clustering oberserved for two different types of initialisations as discussed in the last example.
i.e. in some cases, the number of clusters that form might be less than anticipated because its highly dependent on the way we set up centroids initially.

## Using Sci-kit Learn

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y  = make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
from sklearn.cluster import KMeans
```

```python

kmeans  = KMeans(n_clusters=5, init='k-means++')
```

```python
print(X)
```

```python
# getting results
kmeans.fit(X)
```

```python
# finding centroids of final clusters
centers = kmeans.cluster_centers_
print(centers)
```

```python
# plotting the final cluster centrids along w the points
plt.scatter(X[:,0],X[:,1])
plt.scatter(centers[:,0],centers[:,1],marker='*',color='black')
plt.show()
```

```python
# getting the resultant predictions i.e. mapping of each point to a cluster
pred = kmeans.labels_
print(pred)
```

```python
# this 
print(y)
```

```python
# Scatter Plot for our model
plt.scatter(X[:,0],X[:,1],c=pred)
plt.scatter(centers[:,0],centers[:,1],marker='*',color='black')
plt.title("Prediction")
plt.show()
```

```python
# Actual Data Scatter Plot 
plt.scatter(X[:,0],X[:,1],c=y)
plt.title("Actual")
plt.show()
```
