# Distribution

### Forming a Standard Normal Distribution Graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Establish mean and standard deviation
neu = 5
sigma = 2

# Form a normal distribution of values around the mean
vals = neu + sigma*(np.random.randn(1000))   # rand() is a random number generator that returns normal distribution
print(vals.shape)
print(vals.max())
print(vals.min())
print(vals)
```

```python
# Plot the normal distribution using a Histogram
plt.hist(vals, bins = 100)
plt.xlabel('Intervals of Data')
plt.ylabel('Frequency of that Data Interval')
plt.show()
```

```python
# Scatter plot of the normal distribution
plt.figure(figsize=(20,10))
x = vals
y = np.zeros(x.shape)
plt.scatter(x, y)
plt.show()
```

```python
# analyze to see if the distribution is normal
vals =  np.round(vals)
z = np.unique(vals, return_counts = True)
print(z)

# its clearly visible that the number 5 has max number of occurences or frequency i.e. 211
```

#### Multivariate Normal Distribution


```python
# Bivariate
mean = np.array([0,0])
covariance = np.array([[1,0],[0,1]])

dist = np.random.multivariate_normal(mean, covariance, 500) # 500 points
print(dist.shape)
plt.scatter(dist[:,0], dist[:,1])
plt.show()

# It is visible that sigmaxx = 1 and sigmayy = 1
# while sigmaxy = 0 and sigmayx = 0, which means x and y values are independent of each other
```

```python
# Bivariate
mean2 = np.array([5,6])
covariance2 = np.array([[1.3,0.2],[0.2,1.1]])

dist2 = np.random.multivariate_normal(mean2, covariance2, 500) # 500 points
print(dist2.shape)
plt.scatter(dist2[:,0], dist2[:,1])
plt.show()

# It is visible that sigmaxx = 1 and sigmayy = 1
# while sigmaxy = 0 and sigmayx = 0, which means x and y values are independent of each other
```
